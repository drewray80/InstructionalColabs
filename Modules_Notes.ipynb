## **What are Modules in Python?**  
Modules are groups of code or scripts that we can reuse by importing it into our program.  
Modules can be code that we have written, **user-defined**, or code that someone else has written.  

We will start by learning about modules that someone else has written.  
(Then we will learn how to write your own modules.)


---


Modules are often referred to as "libraries" or "packages".  
The term library is more frequently used in other languages, and in python most people use the word module.  
Packages actually refers to a group of modules, so that is technically different (but some people use the terms interchangably).  


---

## **Usinge Modules**

### **import**  
To be able to access code that is not in our program, we need to import it.  
If you have used an **import** statement at the top of your code, you have used a module.  
  
The **import** keyword will let us import any modules that are built-in to python, installed in our python interpreter, or are currently in the same directory as our program.

After we import the module, we can use any of the objects that are in the module.

Some modules have lots of functions available. To find out what is available, we can look at the module's documentation (by Googling it).  
[random module docs](https://docs.python.org/3.11/library/random.html#module-random)

### **Import Only What You Need**  
As you can see, even a built-in module like random can have a lot of code inside it.  
For this reason, it is a best practice to only import the parts of the module that you need.  
We do this by using:  
**from {module name} import {needed part}**  

This reduces the overall size of our program, and makes our program initialize (load) faster.  


---

You can import all, if you need to.  
The short cut for this is the * symbol.  
Ex: from random import *  
(But this should be avoided.)

### Module Abbreviations  
When we use modules, we have to reference the module name when we call a function.  
Ex:  
random.randint(1, 5)  
or  
date.today()  



---

It can save time and look cleaner to use abbreviations for module names.  
We set these abbreviations in the import statement using as.  
Ex:
import pandas as pd



```python
# If we import a module like pyplot from matplotlib without an abbreviation,
# we will need to type pyplot over and over.

# from matplotlib import pyplot

# pyplot.plot([1, 2, 3, 4])
# pyplot.ylabel('some numbers')
# pyplot.show()
```

```python
# If we write the same code using plt as the abbreviation,
# our code is quicker to write and we will have less spelling errors.

# from matplotlib import pyplot as plt

# plt.plot([1, 2, 3, 4])
# plt.ylabel('some numbers')
# plt.show()
```

The use of abbreviations for modules is so common that some modules have abbreviations that everyone uses.   
Ex:
*   pd  &emsp; pandas  &emsp;&emsp; import pandas as pd

*   sns	&emsp;seaborn	&emsp;&emsp;import seaborn as sns  
*   tt	&emsp;theano.tensor	&emsp;&emsp;import theano.tensor as tt  
*   tf	&emsp;tensorflow	&emsp;&emsp;import tensorflow as tf  
*   pkl	&emsp;pickle	&emsp;&emsp;import pickle as pkl  
*   dt	&emsp;datetime	&emsp;&emsp;import datetime as dt
