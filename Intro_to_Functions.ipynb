# Introduction to Functions

Functions allow us to group code into reusable blocks
that we can call as needed.
Functions reduce the need to rewrite portions of code
and allow us to call code blocks when we need it.

## Defining a Function

def is the keyword that indicates a function is being defined  
After def the function is named, followed by ( ):  
Just like loops, the next lines of code must be indented and
they are the lines that preform the action.



```python
def greeting():
  print("Hello")
#Functions are called by typing the name of the function without indention.

greeting()
```

## Parameters and Arguments  
**Parameters** are inputs that the function uses.  
They are identified in the ( ) when the function is defined.  
def greeting(**name**):  
&emsp; print("Hello " + name)

**Arguments** are inputs that are passed into the function
when it is called.   
greeting(**'Ray'**)

```python
def greeting(name):
  print("Hello " + name)

greeting('Ray')
greeting('Sara')
```

### Multiple parameters can be given.  
When the function is called the arguments use the same order as the parameters.  
You can have as many parameters as you need.

```python
def greeting(first_name, last_name):
  print('Hello ' + first_name + ' ' + last_name)

greeting('Kevin', 'Philips')
greeting('Kim', 'Jones')
```

## Types of Arguments  

### Positional arguments:  
match the position (order) of the parameters of the function  
(the order matters)

```python
def greeting(first_name, last_name):
  print('Hello ' + first_name + ' ' + last_name)

greeting('Kim', 'Jones')
```

### Keyword arguments:  
are identified by the name of the parameter and = (the order doesn't matter)

```python
# This prints the same message because the arguments are set = to the parameter keyword.

greeting(last_name='Jones', first_name='Kim')
greeting()
```

### Default arguments:  
values are set when the function is defined; If the argument is not given, the default will be used.

```python
# If you use positional and keyword arguments in the same function, the positional arguments must be given 1st.

def greeting(first_name = 'Keith', last_name = 'Smith'):
  print('Hello ' + first_name + ' ' + last_name)
greeting()
greeting('Kim')
greeting(last_name='Ray')
```

## Variable Access  

**Scope:** a variable is only available inside the region it is
created.  
**Local Scope:** If a variable in defined inside a function, it
can only be accessed in that function.  
**Global Scope:** If a variable is created in the main body of
the program, it can be accessed anyway in the program

## Returns  

**return** can be used in a function to get a value that can be
used or modified later

This is usually used by assigning the function to a variable.

```python
def profit(income, expenses):
  return income - expenses
concert_profit = profit(5000, 3000)
print(concert_profit)
```
